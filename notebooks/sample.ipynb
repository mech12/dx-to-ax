## 2. 기본 라이브러리 테스트

````python
# NumPy 테스트
try:
    import numpy as np
    print(f"NumPy 버전: {np.__version__}")
    print(f"NumPy 배열 예제: {np.array([1, 2, 3, 4, 5])}")
except ImportError:
    print("NumPy가 설치되어 있지 않습니다. 설치: conda install numpy -y")
````

````python
# Pandas 테스트
try:
    import pandas as pd
    print(f"Pandas 버전: {pd.__version__}")
    df = pd.DataFrame({'A': [1, 2, 3], 'B': [4, 5, 6]})
    print("\nPandas DataFrame 예제:")
    print(df)
except ImportError:
    print("Pandas가 설치되어 있지 않습니다. 설치: conda install pandas -y")
````

````python
# Matplotlib 테스트
try:
    import matplotlib.pyplot as plt
    import matplotlib.font_manager as fm
    print(f"Matplotlib 버전: {plt.matplotlib.__version__}")
    
    # 한글 폰트 설정 (macOS)
    plt.rcParams['font.family'] = 'AppleGothic'  # macOS 기본 한글 폰트
    plt.rcParams['axes.unicode_minus'] = False  # 마이너스 기호 깨짐 방지
    
    # 간단한 그래프
    x = [1, 2, 3, 4, 5]
    y = [1, 4, 9, 16, 25]
    plt.figure(figsize=(8, 6))
    plt.plot(x, y, marker='o')
    plt.title('샘플 그래프')
    plt.xlabel('X축')
    plt.ylabel('Y축')
    plt.grid(True)
    plt.show()
except ImportError:
    print("Matplotlib이 설치되어 있지 않습니다. 설치: conda install matplotlib -y")
````

## 3. 환경 변수 확인

````python
import os

print("현재 작업 디렉토리:", os.getcwd())
print("\nConda 환경 정보:")
conda_env = os.environ.get('CONDA_DEFAULT_ENV', '없음')
print(f"활성화된 환경: {conda_env}")
````

## 4. 설치된 패키지 목록

````python
# 설치된 패키지 목록 확인
import pkg_resources

installed_packages = [(d.project_name, d.version) for d in pkg_resources.working_set]
installed_packages.sort()

print(f"총 {len(installed_packages)}개의 패키지가 설치되어 있습니다.\n")
print("주요 패키지:")
for name, version in installed_packages[:10]:
    print(f"  - {name}: {version}")
````
